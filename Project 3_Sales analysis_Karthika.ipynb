## Sales Analysis

#### Import necessary libraries

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./SalesAnalysis/Sales_Data/Sales_April_2019.csv")
df.head()
```

```python
files = [file for file in os.listdir('./SalesAnalysis/Sales_Data')]
all_data=pd.DataFrame()
for file in files:
    df=pd.read_csv("./SalesAnalysis/Sales_Data/"+file)
    all_data=pd.concat([all_data,df])
all_data.to_csv('./SalesAnalysis/Sales_Data/all_data.csv', index=False)
```

```python
all_data=pd.read_csv("./SalesAnalysis/Sales_Data/all_data.csv")
all_data.head()
```

```python
all_data=all_data.dropna()
```

```python
all_data
```

```python
all_data
```

```python
all_data["Month"]=all_data["Order Date"].str[0:2]
all_data
```

```python
all_data
```

```python
na_df=all_data[all_data.isna().any(axis=1)]
na_df.head()
```

```python
all_data=all_data[all_data["Month"] != "Or"]
```

```python
all_data["Month"]=all_data["Month"].astype("int32")
```

```python
all_data.head()
```

## Best Month sale? and amount of sale

```python
all_data["Quantity Ordered"]=all_data["Quantity Ordered"].astype("int32")
```

```python
all_data["Price Each"]=all_data["Price Each"].astype("float64")
```

```python
all_data["Sales"]=all_data["Quantity Ordered"]*all_data["Price Each"]
all_data
```

```python
sum=all_data.groupby("Month").sum()
sum
```

```python
sum["Sales"].max()
```

```python
months=range(1,13)
plt.bar(months,sum["Sales"])
plt.xticks(months)
plt.ylabel("Sales in USD")
plt.xlabel("Month")
plt.show()
```

#### What time should we display advertisements to maximize the sales?



```python
all_data["Order Date"]=pd.to_datetime(all_data["Order Date"])
all_data
```

```python
all_data["Hour"]=all_data["Order Date"].dt.hour
all_data
```

```python
hours= [hour for hour, df in all_data.groupby("Hour")]
plt.plot(hours, all_data.groupby("Hour").count(), color="red")
plt.xticks(hours)
plt.xlabel("Hours")
plt.ylabel("No of Orders")
plt.grid()
plt.show()
```

### What products are sold together most often?

```python
all_data
```

```python
dup=all_data[all_data["Order ID"].duplicated(keep=False)]
dup["Grouped"]=dup.groupby("Order ID")["Product"].transform(lambda x: ",".join(x))
dup=dup[["Order ID", "Grouped"]].drop_duplicates()
dup.head(20)
```

```python
dup.head(20)
```

#### import more libraries

```python
from itertools import combinations
from collections import Counter
count = Counter()
for row in dup["Grouped"]:
    row_list=row.split(",")
    count.update(Counter(combinations(row_list,2)))
for key, value in count.most_common(10):
    print(key, value)
```


### What product sold the most? and why?

```python
product_count=all_data.groupby("Product")
Quantity_ordered=product_count.sum()["Quantity Ordered"]
products=[product for product,dup in product_count]
plt.bar(products, Quantity_ordered)
plt.xticks(products, rotation="vertical", size=8)
plt.xlabel("Product Name")
plt.ylabel("Number of orders")
```

```python
prices=all_data.groupby("Product").mean()["Price Each"]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, Quantity_ordered, color='green')
ax2.plot(products, prices , 'b')

ax1.set_xticklabels(products, rotation='vertical')
ax1.set_xlabel('Name of the product')
ax1.set_ylabel('Quantity ordered', color='g')
ax2.set_ylabel('Price of the product', color='b')

plt.show()
```
